# Dataset

## TensorFlow

```python
import lulc

# Dataset info.
split_train_val = 0.3
shuffle = True
ratio_val_test = 3
path_dir = 'datasets/Sentinel2GlobalLULC/Sentinel2LULC_JPEG/'    # /home/sfandres/Downloads/Sentinel-dataset/'

(train_dataset, validation_dataset, test_dataset), class_names = \
    lulc.tf_data_loader(path_dir=path_dir,
                        split_train_val=split_train_val,
                        shuffle=shuffle,
                        ratio_val_test=ratio_val_test
                       )
```

```python
import tensorflow_addons as tfa
from tensorflow_addons.metrics import F1Score

# Dealing with the different classes.
n_classes = len(class_names)

def f1_score_macro():
    return tfa.metrics.F1Score(num_classes=n_classes,
                               average='macro',
                               name='f1_score_macro')
```

```python
# Load again the model.
model_path = './my_logs/run-2022_09_23-08_45_58-ResNet50-Data_aug_False/ResNet50-epoch-17-loss-0.101-val_loss-0.108-val_acc-0.956.h5'
reconstructed_model = tf.keras.models.load_model(model_path,
                                                 custom_objects={'F1Score': f1_score_macro})

# Reconstructed model summary.
reconstructed_model.summary()
```

```python
target_dataset = test_dataset

for images, labels in target_dataset.take(1):    # only take first element of dataset, batch in this case as it is prefetched
    
    # Batch id.
    idx = 2
    numpy_batch_images = images.numpy()
    numpy_batch_labels = labels.numpy()
    
    # Plot target img.
    plt.imshow(numpy_batch_images[idx].astype("uint8"))
    print(class_names[np.argmax(numpy_batch_labels[idx])])
    
    # Prediction.
    predictions = reconstructed_model.predict(numpy_batch_images)[idx]
    print('\n', predictions)
    print(class_names[np.argmax(predictions)], predictions[np.argmax(predictions)])
```

# PyTorch

```python
import torch
from torchvision import transforms, datasets
from torch.utils.data import DataLoader

transform = transforms.Compose([
    transforms.ToTensor()
    ])

full_dataset = datasets.ImageFolder(root='datasets/Sentinel2GlobalLULC/Sentinel2LULC_JPEG/',
                                    transform=transform)

train_size = int(0.7 * len(full_dataset))
test_size = len(full_dataset) - train_size

train_dataset, test_dataset = torch.utils.data.random_split(full_dataset, [train_size, test_size])

train_dataloader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_dataloader = DataLoader(test_dataset, batch_size=32, shuffle=True)
```

## Show examples from some random batch

### Iter and next

```python
# Display image and label.
train_features, train_labels = next(iter(train_dataloader))
print(f"Feature batch shape: {train_features.size()}")
print(f"Labels batch shape: {train_labels.size()}")
img = train_features[0].squeeze()
label = train_labels[0]
plt.imshow(img.permute(1, 2, 0), cmap="gray")
plt.show()
print(f"Label: {label}")
```

### Loop

```python
# Accessing Data and Targets in a PyTorch DataLoader
for idx, (image, label) in enumerate(train_dataloader):
    img = image[0].squeeze()
    label = label[0]
    plt.title("Label: " + str(int(label)))
    plt.imshow(img.permute(1, 2, 0), cmap="gray")
    plt.show()
    break
```
